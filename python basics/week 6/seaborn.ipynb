### Seaborn:-

Seaborn is a data visualization library built on top of Matplotlib.

Used to create beautiful, statistical graphs with less code.

- Why Seaborn?

Works directly with Pandas DataFrames

Better default styles & colors

Easy statistical plots

Less code than Matplotlib

- Different categories of plot in Seaborn 

Plots are basically used for visualizing the relationship between variables. Those variables can be either completely numerical or a category like a group, class, or division. Seaborn divides the plot into the below categories - 
 

Relational plots:   
This plot is used to understand the relation between two variables.

Categorical plots:   
This plot deals with categorical variables and how they can be visualized.

Distribution plots:   
This plot is used for examining univariate and bivariate distributions

Regression plots:   
The regression plots in Seaborn are primarily intended to add a visual guide that helps to emphasize patterns in a dataset during exploratory data analyses.

Matrix plots:   
A matrix plot is an array of scatterplots.

Multi-plot grids:   
It is a useful approach to draw multiple instances of the same plot on different subsets of the dataset.

🔸 scatterplot:

👉 Shows relation between two numeric values

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips") # Load the "tips" dataset from seaborn
df
sns.scatterplot(x="total_bill", y="tip", data=df)
plt.show()
```

🔸 lineplot:

👉 Trend over time / order

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.lineplot(x="size", y="tip", data=df)
plt.show()
```

🔸 relplot (figure-level)

```python
sns.relplot(x="total_bill", y="tip", kind="line", data=df) # Create a relational plot with line kind(line, scatter)
```

2️⃣ Categorical Plots (Category-wise comparison):

🔸 barplot:  
👉 Shows mean value

```python
sns.barplot(x="day", y="total_bill", data=df)
plt.show()
```

🔸 countplot:

👉 Counts frequency

```python

sns.countplot(x="day", data=df)
plt.show()
```

🔸 boxplot

👉 Shows:

min : Minimum value 

max :Maximum value

median: middle value

outliers :unusual / far values

```python
sns.boxplot(x="day", y="total_bill", data=df)
plt.show(max)
```

🔸 violinplot

👉 Boxplot + density

| Graph       | Kay Show Karto                  |
| ----------- | ------------------------------- |
| Histogram   | Frequency                       |
| Boxplot     | Min, max, median, outliers      |
| Violin Plot | Distribution + density + median |


- Middle line → median  
- Width → data density   
- jithe width jast tithe values jast ahet

```python
# Numeric data cha distribution show karto
# Boxplot + Density plot combination ahe
# Wide part → data density जास्त
# Narrow part → data density कमी
# White dot → median
# Thick bar → IQR (Interquartile Range) # 25th percentile to 75th percentile
# Thin lines → remaining data spread
# Outliers pan identify hotat


sns.violinplot(x="day", y="total_bill", data=df)
plt.show()
```

🔸 stripplot

```python
# Strip plot is a scatter plot where one of the variables is categorical. 
# It is used to show the distribution of a categorical variable and its relationship with a continuous variable.

# Individual values show karto
# Data spread samajto
# Outliers easily दिसतात
# Mostly categorical + numerical data sathi use hoto
# Overlapping points adjust karat nahi
# jitter parameter use karun points thoda shift karu shakto

sns.stripplot(x="day", y="total_bill", data=df, jitter=True)
plt.show()
```

🔸 swarmplot

```python
# Swarm plot is a scatter plot where one of the variables is categorical. 
# It is used to show the distribution of a categorical variable and its relationship with a continuous variable. 
# It is similar to a strip plot, but the points are adjusted so that they do not overlap.

# Individual data points display karto
# Overlapping avoid karto
# Data distribution clearly दिसतो
# Outliers easily identify hotat
# Strip plot cha improved version samajla jato

sns.swarmplot(x="day", y="total_bill", data=df)
plt.show()
```

🔸 catplot (figure-level)

kind= use karun konta categorical graph pahije te select karto.

Example:

- strip
- swarm
- box
- violin

```python
# A boxen plot is a type of box plot that is used to show the distribution of a continuous variable. 
# It is similar to a box plot, but it shows more information about the distribution of the data, such as the density of the data points.

sns.catplot(x="day", y="total_bill", kind="boxen", data=df)
```

3️⃣ Distribution Plots (Data spread)

🔸 histplot

Seaborn Histplot is used to visualize the distribution of a single variable (univariate data).  
It shows how data values are spread across intervals (bins).

```python
sns.histplot(df["total_bill"], bins=5) # Create a histogram of the "total_bill" column with 5 bins
plt.show()
```

🔸 kdeplot

👉 Smooth curve (density)

Shows Kernel Density Estimate

Smooth curve over histogram

Default is False

```python
sns.kdeplot(df["total_bill"])
plt.show()
```

🔸 displot

Seaborn distplot() is used to visualize the distribution of a single variable (univariate data).

```python
sns.displot(df["total_bill"], kind="hist")
```

4️⃣ Regression Plots (Trend + relationship):

🔸 regplot

👉 Scatter + regression line

```python
# A regression plot is a type of plot that shows the relationship between two variables. 
# It is used to visualize the relationship between a dependent variable and an independent variable.
# It is a combination of a scatter plot and a line plot, where the scatter plot shows the individual data points and 
# the line plot shows the best fit line for the data.

sns.regplot(x="total_bill", y="tip", data=df) 
plt.show()
```

🔸 Lmplot

```python
# lmplot is a higher-level interface for drawing regression plots. It combines regplot() and FacetGrid.


sns.lmplot(x="total_bill", y="tip", data=df)
```

5️⃣ Matrix Plots (Correlation & Heatmaps):

🔸 heatmap

👉 Very important for ML & Data Analysis

```python
# Correlation Matrix :
# A correlation matrix shows the relationship between variables

# Correlation matrix chi heatmap banvaychi asel tar
# visualization on correlation matrix

# 1 → strong positive relation
# -1 → strong negative relation
# 0 → no relation

corr = df.corr(numeric_only=True) # Calculate the correlation matrix for the numeric columns in the DataFrame
sns.heatmap(corr, annot=True, cmap="coolwarm") # Create a heatmap of the correlation matrix with annotations and a coolwarm color map
plt.show()
```

6️⃣ Multi-Plot Grids

🔸 pairplot

👉 Plots relationships between all numeric columns

```python
sns.pairplot(df)
```

🔸 facetgrid

```python
# FacetGrid is a class in the seaborn library that allows you to create a grid of subplots based on the values of one or more categorical variables. 
# It is used to visualize the relationship between multiple variables in a dataset by creating a grid of plots that are organized by the values of the categorical variables.

g = sns.FacetGrid(df, col="day")
g.map(sns.histplot, "total_bill")
```

#### - Styling & Themes 

🔸 Set theme

Styles:

darkgrid

whitegrid

dark

white

ticks

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.lineplot(x="size", y="tip", data=df)
sns.set_theme(style="darkgrid")
plt.show()
```

🔸 Color palettes

Popular palettes:

deep

muted

bright

pastel

Set1, Set2

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("tips")
sns.lineplot(x="size", y="tip", data=df)
sns.color_palette("pastel")
sns.set_palette("Set2") # set 1,2,3 ..... colors
plt.show()
```

#### - Hue, Size & Style 

👉 Multiple info in one plot

```python
sns.scatterplot(
    x="total_bill", # x-axis var
    y="tip", # y-axis var
    hue="sex", # color by
    size="size", # size by
    style="time", # style by
    data=df
)
plt.show()
```
