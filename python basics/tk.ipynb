### Tkinter:-

-Tkinter is Pythonâ€™s built-in library for creating graphical user interfaces (GUIs).   
-It acts as a lightweight wrapper around Tcl/Tk GUI toolkit, offering Python developers a simple and intuitive way to build    desktop applications.   
-It supports layout management, event handling and customization, making it ideal for rapid GUI development in Python.  
-Tkinter is Pythonâ€™s standard GUI library used to create desktop applications.

- Importing Tkinter:

  import tkinter as tk

	or

  from tkinter import *

- Main Window (Root Window)

root = tk.Tk()  
root.mainloop()

Method      | Purpose
------------|-----------------------
Tk()	      | Creates main window
mainloop()  | Keeps window running


- Window Properties:


root.title("My App")  
root.geometry("400x300")  
root.resizable(False, False)

Property	  | Use
------------|------------------
title()	    | Window title
geometry()	| Width x Height
resizable()	| Disable resize

```python
import tkinter as tk

# create window
window = tk.Tk()
window.title("My First Tkinter App")
window.geometry("300x200")  #...(Width, Height)

# create label
label = tk.Label(window, text="Hello Tkinter!", font=("Arial", 14))
label.pack(pady=40)

# run window
window.mainloop()   #....The mainloop() method is used to run application once it is ready.
```

### 1. Label
It refers to the display box where we display text or image. It can have various options like font, background, foreground, etc.

Syntax:

    w=Label(master, option=value)

```python
import tkinter as tk

# create window
window = tk.Tk()
window.title("My First Tkinter App")
window.geometry("300x200")  #...(Width, Height)

# create label
label = tk.Label(window, text="Hello Tkinter!", font=("Arial", 14))
label.pack(pady=5)

# run window
window.mainloop()
```

### 2. Button
A clickable button that can trigger an action.

Syntax:

    w=Button(master, option=value)

```python
import tkinter as tk

window = tk.Tk()
window.title("Button Example")
window.geometry("300x200")

button=tk.Button(window, text='Stop', width=25, command=window.destroy)  #.....destroy() is optional
# It is used only when you want to close the window by code
button.pack(pady=20) 

window.mainloop()
```

### 3. Entry
It is used to input the single line text entry from the user. For multi-line text input, Text widget is used.

### 4. CheckButton
A checkbox can be toggled on or off. It can be linked to a variable to store its state.

Syntax:

    w = CheckButton(master, option=value)

```python
import tkinter as tk

window = tk.Tk()
window.title("Checkbutton Example")
window.geometry("300x200")

var1 = tk.IntVar()
var2 = tk.IntVar()

cb1 = tk.Checkbutton(window, text="Male", variable=var1)
cb1.grid(row=0, sticky=tk.W)

cb2 = tk.Checkbutton(window, text="Female", variable=var2)
cb2.grid(row=1, sticky=tk.W)

window.mainloop()
```

### 5. RadioButton
It allows user to select one option from a set of choices. They are grouped by sharing the same variable.

Syntax:

    w = RadioButton(master, option=value)

```python
import tkinter as tk

window = tk.Tk()
window.title("Radiobutton Example")
window.geometry ("400x300") 

gender = tk.StringVar()
gender.set("Male")   # default selection

rb1 = tk.Radiobutton(window, text="Male", variable=gender, value="Male")
rb1.grid(row=0, sticky=tk.W)  #....sticky=tk.W aligns them left

rb2 = tk.Radiobutton(window, text="Female", variable=gender, value="Female")
rb2.grid(row=1, sticky=tk.W)

window.mainloop()
```

### 7. Scrollbar
It refers to the slide controller which will be used to implement listed widgets.

Syntax:

    w = Scrollbar(master, option=value)

```python
import tkinter as tk

window = tk.Tk()
window.title("Scrollbar Example")
window.geometry("300x250")

# Create Listbox
lb = tk.Listbox(window, height=8)
lb.pack(side=tk.LEFT, fill=tk.BOTH)

# Create Scrollbar
sb = tk.Scrollbar(window)
sb.pack(side=tk.RIGHT, fill=tk.Y)

# Connect scrollbar and listbox
lb.config(yscrollcommand=sb.set)
sb.config(command=lb.yview)  #....lb.yview â†’ scrolls listbox vertically

# Insert many items
for i in range(1, 31):
    lb.insert(tk.END, f"Item {i}")

window.mainloop()
```

### 8. Menu
It is used to create all kinds of menus used by the application.

Syntax:

    window.w = Menu(master, option=value)

```python
import tkinter as tk

window = tk.Tk()
window.title("Menu Example")
window.geometry("300x200")

# Create menu bar
menubar = tk.Menu(window)
window.config(menu=menubar)

# Create File menu
filemenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label="File", menu=filemenu)

filemenu.add_command(label="New")  #...add_command() â†’ adds menu item
filemenu.add_command(label="Open")
filemenu.add_separator()
filemenu.add_command(label="Exit", command=window.destroy)

# Create Help menu
helpmenu = tk.Menu(menubar, tearoff=0)
menubar.add_cascade(label="Help", menu=helpmenu)

helpmenu.add_command(label="About")

window.mainloop()
```

### 9. Combobox
The Combobox widget from tkinter.ttk is created using Combobox class. Its options are set via values parameter, and a default value is assigned using set() method. An event handler (like on_select) can be bound using bind() to update other widgets based on selected item.

Syntax:

    combo = Combobox(master, values=[...], state='readonly')

```python
import tkinter as tk
from tkinter import ttk   #...ttk provides modern widgets  // Combobox is not in normal Tkinter, it is in ttk

window = tk.Tk()
window.title("Combobox Example")
window.geometry("300x200")

# Create Combobox
course = ttk.Combobox(window)
course['values'] = ("Python", "Java", "C++", "Data Science")

course.current(0)   # default selection
course.pack(pady=20)

window.mainloop()
```

### 13. MenuButton
It is a part of top-down menu which stays on the window all the time. Every menubutton has its own functionality.

Syntax:

    w = MenuButton(master, option=value)

```python
import tkinter as tk

top = tk.Tk()
top.title("Menu button example")
top.geometry("300x200")

mb = tk.Menubutton(top,text="Vaishnavi",relief=tk.RIDGE, bd=2)  # border stylebd=3 / border thickness
mb.pack(pady=30)

mb.menu = tk.Menu(mb, tearoff=0)
mb["menu"] = mb.menu

cVar = tk.IntVar()
aVar = tk.IntVar()

mb.menu.add_checkbutton(label="Contact", variable=cVar)
mb.menu.add_checkbutton(label="About", variable=aVar)

top.mainloop()
```

### 14. Progressbar
progressbar indicates the progress of a long-running task. When the button is clicked, the progressbar fills up to 100% over a short period, simulating a task that takes time to complete.

Syntax:

    Progressbar(parent, orient, length, mode)

```python
import tkinter as tk
from tkinter import ttk
import time

window = tk.Tk()
window.title("Progress Bar Example")
window.geometry("300x150")

# Create Progressbar
progress = ttk.Progressbar(
    window,
    orient=tk.HORIZONTAL,
    length=250,
      mode='determinate'
)
progress.pack(pady=30)

# Start progress
for i in range(51):
    progress['value'] = i
    window.update_idletasks()  #...update_idletasks():- refreshes UI
    time.sleep(0.03)

window.mainloop()
```
