### File Handling

Python has several functions for creating, reading, updating, and deleting files.  
This is important when you want to store the data permanently or work with large datasets.  
The key function for working with files in Python is the open() function.  
The open() function takes two parameters; filename, and mode.

- There are four different methods (modes) for opening a file:

"r" - Read - Default value. Opens a file for reading, error if the file does not exist  
"a" - Append - Opens a file for appending, creates the file if it does not exist  
"w" - Write - Opens a file for writing, creates the file if it does not exist  
"x" - Create - Creates the specified file, returns an error if the file exists

- In addition you can specify if the file should be handled as binary or text mode 
 
"t" - Text - Default value. Text mode  
"b" - Binary - Binary mode (e.g. images)

-----------------------------------------------------------------------------------------------

- Open a File

```python
f= open('abc.txt','r')
```

- Read from a file

Once a fie is open, you can read from it using the following methods:

read(): Reads the entire content of the file.  
readline(): Read one line from the file at a time.  
readliness(): Read all lines into a list.

```python
# Read():
f = open('abc.txt','r')
content=f.read()
print(content)
f.close()
```

```python
# readline():
f=open('abc.txt','r')
content=f.readline()
print(content)
f.close()
```

```python
# readliness(): 
f=open('abc.txt','r')
content=f.readlines()
print(content)
f.close
```

-  Write to file

To write to a file, you can use the write() or writelins() method:

write(): Write a string to the file.  
writelines(): Write the list of string.

```python
# write():

f=open ('abc.txt','w')
f.write("Hello World!!")
f.close()
```

- append:

Append means adding new data at the end of an existing file.  
Old data is not deleted

```python
# append mode: incremental add / write

f= open('abc.txt','a')
f.write("\nVaishnavi Jadhav")
f.close()
```

```python
# writelines():

f=open ('abc.txt','w')
content= f.writelines(["Python"," Java"," OOP"])
print(content)
f.close()
```

- using with statement:  

    Automatically closes file

```python
with open('abc.txt','r') as f:
    content= f.read()
    print(content)
```

```python
## File Handling with Exception

try:
    with open('cde.txt','r') as f:
        print(f.read())
except FileNotFoundError:
    print("Flie not found!!")
```
