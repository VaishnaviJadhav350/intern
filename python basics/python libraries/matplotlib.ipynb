#### - Pyplot
Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:  
    
    import matplotlib.pyplot as plt

```python
## Draw a line in a diagram from position (1, 15) to position (4, 35):

import matplotlib.pyplot as plt

x = [1, 15]
y = [4, 35]

plt.plot(x, y)
plt.show()    # optional
```

#### - plotting:

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.(x-axis: horizontal)

Parameter 2 is an array containing the points on the y-axis.(y-axis: vertical)

```python
## Draw a line in a diagram from position (1, 3) to position (8, 10):

import matplotlib.pyplot as plt

x=[1, 3]
y=[8, 10]

plt.plot(x, y)
plt.show()    # optional    
```

#### - Markers
You can use the keyword argument marker to emphasize each point with a specified marker

You can choose any of these markers:

'*'	    |    Star	
'.'	    |    Point	
','	    |    Pixel

```python
## Mark each point with a circle:

import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]
xpoints=[0.3, 1.3, 2.3, 3.3]
plt.plot(xpoints, ypoints, marker = '|')
plt.show()
```

-Marker Size:

You can use the keyword argument markersize or the shorter version, 'ms' to set the size of the markers  

    marker size = ms

-Marker Edge colour:

You can use the keyword argument markeredgecolor or the shorter 'mec' to set the color of the edge of the markers

    markeredgecolor = mec


```python
import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]

plt.plot(ypoints, 'o-', ms=10, mec='red') # red edge circle markers with solid line
plt.show()
```

-Marker Face Colour:

You can use the keyword argument markerfacecolor or the shorter 'mfc' to set the color inside the edge of the markers

    MarkerFaceColor = mfc

```python
import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]

plt.plot(ypoints, 'o-', ms=15, mfc='yellow') # yellow filled circle markers with solid line
plt.show()
```

#### - Line:

You can use the keyword argument linestyle, or shorter 'ls', to change the style of the plotted line

    linestyle = ls

Style	  |     Or
----------|---------------
'solid'   | 	'-'	
'dotted'  |      ':'	
'dashed'  |	    '--'	
'dashdot' |	    '-.'	
'None'	  |      'or'


```python
import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]
xpoints=[0, 1, 2, 3]

plt.plot(ypoints, ls='--')
plt.plot(xpoints, ls='dotted')
plt.show()
```

-Line Color:

You can use the keyword argument color or the shorter 'c' to set the color of the line

    color = c

```python
import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]
xpoints=[0, 1, 2, 3]

plt.plot(ypoints, c='green')
plt.plot(xpoints, color="#AF784C")
plt.show()
```

-Line Width:

You can use the keyword argument linewidth or the shorter 'lw' to change the width of the line.

The value is a floating number

    linewidth = lw

```python
import matplotlib.pyplot as plt

ypoints =[3, 8, 1, 10]

plt.plot(ypoints, lw= 15.4)
plt.show()
```

#### - Labels:

With Pyplot, you can use the xlabel() and ylabel() functions to set a label for the x- and y-axis.

    set the label = xlabel, ylebel  
    set the title = tittle

```python
## Add labels to the x- and y-axis:

import matplotlib.pyplot as plt

x = [80, 85, 90, 95, 100, 105, 110, 115, 120, 125]
y = [240, 250, 260, 270, 280, 290, 300, 310, 320, 330]

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

-Set Font Properties:

You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

-Position the Title:

You can use the loc parameter in title() to position the title.

Legal values are:

    'left', 'right', and 'center'. Default value is 'center'.

```python
## Set font properties for the title and labels:

import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':17}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1, loc = 'left')  # tittle + position
plt.xlabel("Average Pulse", fontdict = font2)  # x-axis label with font properties
plt.ylabel("Calorie Burnage", fontdict = font2) # y-axis label with font properties

plt.plot(x, y, ':r')
plt.show()
```
