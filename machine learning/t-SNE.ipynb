## t-SNE:   
#### (t-Distributed Stochastic Neighbor Embedding)

-SNE is a machine learning technique used for visualization.  
-SNE is used to convert high-dimensional data into 2D/3D for visualization while keeping similar data points close together.

High-dimensional data (ज्यात खूप features असतात) 2D किंवा 3D madhye convert karto, so that apan easily visualize karu shakto.

#### Simple Concept:

Samja tumchya kade data aahe with many features (like 50–100).

Direct visualize karta yet nahi ❌  
t-SNE kay karto 👉  
Similar points la जवळ ठेवतो ani dissimilar points la दूर ठेवतो

👉 Marathi madhye:

ज्या data points ekmekanshi similar aahet te जवळ दिसतात
ani je different aahet te दूर दिसतात

#### Example:

#### Student Performance Visualization

👉 Samja aplyakade student data aahe (marks in subjects)  
👉 t-SNE use karun apan 2D graph madhye visualize karu

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

# Sample student data
data = {
    'Maths':   [85, 78, 90, 60, 72, 88, 95, 55],
    'Science': [80, 70, 92, 65, 75, 85, 98, 60],
    'English': [78, 65, 88, 70, 68, 82, 90, 58],
    'Grade':   ['A','B','A','C','B','A','A','C']
}

df = pd.DataFrame(data)

# Features (marks)
X = df[['Maths', 'Science', 'English']]

# t-SNE apply
tsne = TSNE(n_components=2, random_state=42, perplexity=3)
X_tsne = tsne.fit_transform(X)

# Plot
plt.scatter(X_tsne[:,0], X_tsne[:,1])

# Label each point
for i, grade in enumerate(df['Grade']):
    plt.text(X_tsne[i,0], X_tsne[i,1], grade)

plt.title("t-SNE Visualization (Student Data)")
plt.xlabel("Dimension 1")
plt.ylabel("Dimension 2")
plt.show()
```
