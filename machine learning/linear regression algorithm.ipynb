#### Split Data

```python
x = df.drop("price", axis=1)
y = df["price"]

print("Shape of x =", x.shape)
print("Shape of y =", y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
print('Shape of x_train =', x_train.shape)
print('Shape of x_test =', x_test.shape)
print('Shape of y_train =', y_train.shape)
print('Shape of y_test =', y_test.shape)
```

#### Feature Scalling

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

#### Accuracy Matrics for Regression 

1Ô∏è‚É£ Mean Absolute Error (MAE)  
2Ô∏è‚É£ Mean Squared Error (MSE)  
3Ô∏è‚É£ Root Mean Squared Error (RMSE)  
4Ô∏è‚É£ R¬≤ Score (Coefficient of Determination)

- Why used Accuracy matrics 

-To measure how good a machine learning model is.

-To check how close predicted values are to actual values.  
Because in regression, outputs are continuous numbers, not right or wrong.

-To compare different models and choose the best one.

-To improve the model‚Äôs performance.

##### 1Ô∏è‚É£ Mean Absolute Error (MAE) 

Average of absolute differences between actual and predicted values.

üëç Pros

Easy to understand  
Same unit as target  
Not highly affected by outliers

üëé Cons

Treats all errors equally

##### 2Ô∏è‚É£ Mean Squared Error (MSE)

Squares the error ‚Üí big errors get heavy penalty.

üëç Pros

Differentiable (good for optimization)  
Penalizes large mistakes

üëé Cons

Units become squared  
Very sensitive to outliers

##### 3Ô∏è‚É£ Root Mean Squared Error (RMSE)

Square root of MSE ‚Üí back to original unit.

üëç Pros

Most commonly used  
Easy to interpret  
Penalizes large errors

üëé Cons

Sensitive to outliers

#### 4Ô∏è‚É£ R¬≤ Score (Coefficient of Determination)

Tells how much variance is explained by the model.

R¬≤ Value|	Meaning
--------|----------------------
1.0	    |    Perfect model
0.7	    |    Good
0.0	    |    No learning
< 0	    |    Worse than mean

üëç Pros

Scale independent  
Easy model comparison

üëé Cons

Increases automatically when features increase
