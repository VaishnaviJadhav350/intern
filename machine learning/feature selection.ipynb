## Feature Selection

Feature Selection is the process of selecting important features (columns) from a dataset and removing irrelevant or unnecessary features.  
We keep only those columns that are useful for the machine learning model and remove the rest.

Feature Selection म्हणजे dataset मधील important features (columns) निवडणे आणि irrelevant / unnecessary features काढून टाकणे.  
Dataset मधून जे columns model साठी उपयोगी आहेत ते ठेवणे आणि बाकीचे remove करणे.

#### Why is Feature Selection Important?

It improves model performance   
It reduces training time   
It helps reduce overfitting  
It makes the model simple and efficient  
The model becomes faster and more accurate.

##### Example:

Student Dataset:  
Name ❌  
Roll No ❌  
Marks ✅  
Study Hours ✅  

👉 Here:  

Marks and Study Hours are useful  
Name and Roll No are not useful

So, we use only the important features for building the model.

#### Types of Feature Selection

1️⃣ Filter Method  
2️⃣ Wrapper Method  
3️⃣ Embedded Method  

#### 1️⃣ Filter Method :

Filter Method is a feature selection technique where we select features using statistical measures, without using any machine learning model.  
Statistical techniques वापरून features select करतो  
Model वापरत नाही  
Data मधील features चा relation check करून selection करतो.

##### Techniques:

1️⃣ Correlation :  

Numeric data साठी वापरतो  
High correlation = important feature

2️⃣ Chi-Square Test :  

Categorical data साठी वापरतो

3️⃣ ANOVA (Analysis of Variance) : 
 
Feature आणि target मधील difference check करतो

##### Example:

जर दोन features same information देत असतील 👉 एक remove करतो

✅ Advantages:  
Fast   
Simple

❌ Disadvantages:  
Model performance consider करत नाही

```python
import pandas as pd

data = {
    'Age': [20, 21, 22, 23, 24],
    'Marks': [60, 65, 70, 75, 80],
    'StudyHours': [2, 3, 4, 5, 6],
    'Random': [10, 5, 3, 8, 2]
}

df = pd.DataFrame(data)

# correlation
corr = df.corr()

# select important features
selected = corr['Marks'][corr['Marks'] > 0.5].index
print(selected)
```

#### 2️⃣ Wrapper Method

Wrapper Method is a feature selection technique where we use a machine learning model to evaluate and select the best combination of features.  
Machine Learning model वापरून features select करतो  
Model run करून best feature combination शोधतो.

##### Techniques:

1️⃣ Forward Selection:

Start with no features  
Add features one by one  
Keep the features that improve performance

2️⃣ Backward Elimination:

Start with all features  
Remove one feature at a time  
Remove the least important features

3️⃣ Recursive Feature Elimination (RFE):

Train the model  
Remove the least important feature  
Repeat the process until the desired number of features is selected 

##### Example:

Dataset:

Age  
Study Hours  
Random  
Marks  

👉 Try combinations:

Age + Study Hours → Good performance  
Age + Random → Poor performance

👉 So, select Study Hours and remove Random

Step by step features add/remove करून best result मिळवतो

✅ Advantages:  
More accurate

❌ Disadvantages:  
Slow   
Computationally expensive

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
import pandas as pd

data = {
    'Age': [20, 21, 22, 23, 24],
    'StudyHours': [2, 3, 4, 5, 6],
    'Random': [10, 5, 3, 8, 2],
    'Marks': [60, 65, 70, 75, 80]
}

df = pd.DataFrame(data)

X = df[['Age', 'StudyHours', 'Random']]
y = df['Marks']

model = LinearRegression()

rfe = RFE(model, n_features_to_select=2)
fit = rfe.fit(X, y)

print("Selected Features:", X.columns[fit.support_])
```

#### 3️⃣ Embedded Method :

The Embedded Method is a feature selection technique where feature selection happens automatically during the model training process.  
Feature selection model training दरम्यानच होते  
Model स्वतः training करताना important features select करतो.

##### Techniques:  

1️⃣ Lasso Regression (L1 Regularization):

Reduces some feature coefficients to zero  
Features with zero importance are removed

2️⃣ Decision Tree:

Automatically selects important features based on splits

3️⃣ Random Forest:

Provides feature importance scores

##### Example:

Dataset:

Age  
Study Hours  
Random  
Marks

👉 Model training:

Age → important  
Study Hours → important  
Random → not important (ignored)

👉 So, the model automatically selects useful features.

Decision Tree automatically important features निवडतो

✅ Advantages:  
Fast + Accurate 
Balanced method

❌ Disadvantages:  
Specific models वर depend असतो

```python
from sklearn.linear_model import Lasso
import pandas as pd

data = {
    'Age': [20, 21, 22, 23, 24],
    'StudyHours': [2, 3, 4, 5, 6],
    'Random': [10, 5, 3, 8, 2],
    'Marks': [60, 65, 70, 75, 80]
}

df = pd.DataFrame(data)

X = df[['Age', 'StudyHours', 'Random']]
y = df['Marks']

model = Lasso(alpha=0.1)
model.fit(X, y)

print("Feature Coefficients:", model.coef_)
```

If a coefficient is 0, that feature is not important.

## Hyperparameter Tuning

Hyperparameter Tuning is the process of selecting the best values for model parameters to improve performance.

Hyperparameter Tuning म्हणजे machine learning model चे settings (parameters) best values ला set करणे, जेणेकरून model ची performance improve होईल.  
Model training आधी आपण जे parameters manually set करतो, त्यांना hyperparameters म्हणतात आणि त्यांना best value देण्याची process म्हणजे hyperparameter tuning.

#### Key Idea:

- Try different values  
- Train the model  
- Select the value that gives the best result  

We test different values and choose the one that produces the best output.

#### Example:

In a KNN model:

K = 1 → low accuracy  
K = 3 → better  
K = 5 → best  

👉 So, we select K = 5 as it gives the best performance.

### Methods of Hyperparameter Tuning:

#### 1️⃣ Grid Search :

Grid Search tries all possible combinations of hyperparameter values.

सगळ्या possible values combinations check करतो

#### Example:

K = 1, 3, 5  
Depth = 2, 4

Combinations:

(1,2), (1,4), (3,2), (3,4), (5,2), (5,4)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

# Load dataset
data = load_iris()
X, y = data.data, data.target

# Model
model = KNeighborsClassifier()

# Hyperparameters
param_grid = {
    'n_neighbors': [1, 3, 5, 7],
    'weights': ['uniform', 'distance']
}

# Grid Search
grid = GridSearchCV(model, param_grid, cv=5) # 5-fold cross-validation
grid.fit(X, y)

print("Best Parameters:", grid.best_params_)
```

#### 2️⃣ Random Search :

Random Search tries random combinations of hyperparameters instead of all.

Random values निवडून test करतो

#### Example:

Randomly tries:

(1,2), (5,4), (3,2)

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
import numpy as np

data = load_iris()
X, y = data.data, data.target

model = KNeighborsClassifier()

param_dist = {
    'n_neighbors': np.arange(1, 10),
    'weights': ['uniform', 'distance']
}

random = RandomizedSearchCV(model, param_dist, n_iter=5, cv=5) # 5 iterations - 5 random combinations, 5-fold cross-validation
random.fit(X, y)

print("Best Parameters:", random.best_params_)
```

#### 3️⃣ Bayesian Optimization :

This method uses previous results to choose the next best set of hyperparameters.

आधीच्या results वरून पुढची best value निवडतो

```python
from skopt import BayesSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris

data = load_iris()
X, y = data.data, data.target

model = KNeighborsClassifier()

param_space = {
    'n_neighbors': (1, 10),
    'weights': ['uniform', 'distance']
}

bayes = BayesSearchCV(model, param_space, n_iter=10, cv=5) # n_iter=10 - 10 iterations of Bayesian optimization, cv=5 - 5-fold cross-validation
bayes.fit(X, y)

print("Best Parameters:", bayes.best_params_)
```

#### 4️⃣ Manual Search (Trial & Error) :

Manually values change करून best result शोधतो

स्वतः values बदलून try करतो

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score

data = load_iris()
X, y = data.data, data.target

for k in [1, 3, 5, 7]:
    model = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(model, X, y, cv=5)
    print(f"K={k}, Accuracy={scores.mean()}") # Average accuracy across 5 folds
```
