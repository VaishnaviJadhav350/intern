## Ensemble Learning

nsemble Learning (ML) म्हणजे multiple machine learning models एकत्र वापरून better result मिळवण्याची technique.

"एक model चुकू शकतो, पण multiple models together वापरले तर accuracy वाढते"

##### Basic Idea

तुम्ही एक decision घेण्यासाठी फक्त एका व्यक्तीला विचारण्याऐवजी 5–10 लोकांना विचारता → मग majority opinion घेतला तर decision better होतो.

➡️ ML मध्ये हेच concept वापरतात = Ensemble Learning

##### Why use Ensemble Learning?

Accuracy improve होते  
Overfitting कमी होते  
Model more stable बनतो  

### Types :

### 1️⃣ Bagging (Parallel Learning)

Bagging is an ensemble technique that reduces variance by training multiple models on different bootstrap samples and combining their predictions.

Bagging ही ensemble learning technique आहे ज्यात  
 multiple models same algorithm वापरून train केले जातात पण वेगवेगळ्या sampled datasets वर, आणि शेवटी त्यांचे outputs combine केले जातात.  

एकच model train केल्यावर तो overfit होऊ शकतो ❌  
पण जर आपण multiple models train केले आणि त्यांचा average घेतला 👉 result stable होतो

- #### Key Concept: Bootstrap Sampling

👉 Bootstrap म्हणजे:

Original dataset मधून random samples घेतो  
with replacement (same data point multiple वेळा येऊ शकतो)

- #### Example

Original Data:

[1, 2, 3, 4, 5]

Bootstrap Samples:

Sample 1 → [1, 2, 2, 4, 5]    
Sample 2 → [3, 3, 1, 5, 4]    
Sample 3 → [2, 4, 4, 5, 1]  

👉 प्रत्येक sample different आहे → models diverse होतात

- #### Why Bagging Works? (Deep Insight)

👉 Main problem: High Variance (Overfitting)

Decision Trees सारखे models training data वर overfit होतात  
Bagging → different datasets वापरतो  
त्यामुळे models independent होतात  

👉 Final averaging → variance कमी

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Dataset load
data = load_iris()
X, y = data.data, data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Bagging Model
model = BaggingClassifier(
    estimator=DecisionTreeClassifier(),
    n_estimators=10
)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

### 2️⃣ Boosting (Sequential Learning)

Boosting is an ensemble technique where models are trained sequentially, and each new model focuses on correcting the errors of the previous ones to reduce bias and improve accuracy.

Boosting ही ensemble learning technique आहे ज्यात  
models sequentially (एकामागोमाग) train होतात आणि प्रत्येक next model previous model च्या चुका (errors) correct करतो.

👉 Bagging मध्ये models independent असतात  
👉 पण Boosting मध्ये:

“Focus on mistakes”

First model → काही errors करतो  
Next model → त्या errors वर जास्त लक्ष देतो  
Next → अजून सुधारणा

➡️ शेवटी strong model तयार होतो

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Dataset
data = load_iris()
X, y = data.data, data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Boosting Model
model = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1),
    n_estimators=50,
    learning_rate=1.0
)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```

### 3️⃣ Stacking (Meta Learning)

Stacking is an ensemble learning technique where multiple base models are combined using a meta-model that learns how to best combine their predictions.

Stacking (Stacked Generalization) ही ensemble learning technique आहे ज्यात:  
multiple different models train केले जातात आणि त्यांचे outputs एका new model (meta-model) ला देऊन final prediction घेतला जातो.

- #### Why Stacking Works? (Deep Insight)

👉 Different models → different patterns learn करतात

SVM → margin based learning  
Tree → rule based  
KNN → distance based  

➡️ Combination → अधिक powerful model

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Dataset
data = load_iris()
X, y = data.data, data.target

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Base models
estimators = [
    ('dt', DecisionTreeClassifier()),
    ('svm', SVC(probability=True))
]

# Stacking Model
model = StackingClassifier(
    estimators=estimators,
    final_estimator=LogisticRegression()
)

# Train
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
```
