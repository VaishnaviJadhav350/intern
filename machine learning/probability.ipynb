## Probability:
#### What is probability?

Probability is a branch of mathematics that measures the chance or likelihood of an event happening.  
👉 Probability tells us how likely something is to happen.

The value of probability always lies between 0 and 1.

- 0 → Impossible event  
- 1 → Certain event  
- Between 0 and 1 → Possible event

##### Example

- Getting Head when tossing a coin  
- Getting 6 on a dice  
- Rain tomorrow

##### Example of coin toss:

    P(Head)= 2/1 
           = 0.5

Meaning: There is a 50% chance of getting head.

#### 1. Weather Forecast

Meteorologists use probability to predict rain.

"60% chance of rain tomorrow"

#### 2. Machine Learning & AI

Probability helps models predict outcomes.

##### Example:

##### Example:

### - Rules Of Probability:

##### 1️⃣ Addition Rule of Probability

The Addition Rule is used when we want the probability that event A OR event B occurs.

Meaning: At least one event happens.

Formula:

P(A∪B)=P(A)+P(B)−P(A∩B)

Where:

P(A) = Probability of event A  
P(B) = Probability of event B  
P(A∩B) = Probability that both A and B happen

We subtract the common part so we do not count it twice.

##### - Example (Dice) :

Dice roll केला.

Event A = Even numbers = {2,4,6}

Event B = Number greater than 4 = {5,6}

Step 1: Probability

P(A) = 3/6

P(B) = 2/6

Common outcome

A ∩ B = {6}

P(A ∩ B) = 1/6

Step 2: Apply formula

P(A∪B) = 3/6 + 2/6 − 1/6

P(A∪B) = 4/6 = 2/3

##### 2️⃣ Multiplication Rule of Probability

The Multiplication Rule is used when we want the probability that event A AND event B occur together.

Meaning: Both events happen.

Formula (Independent Events)

P(A∩B)=P(A)×P(B)

##### Example (Two Coins) :

2 coins toss केले.

Find probability of HH (two heads).

Step 1

P(H) = 1/2

Step 2

P(HH)=1/2×1/2

P(HH)=1/4

म्हणजे two heads येण्याची probability = 1/4

| Addition Rule           | Multiplication Rule    |
| ----------------------- | ---------------------- |
| Used for **OR (किंवा)** | Used for **AND (आणि)** |
| A किंवा B               | A आणि B                |
| Formula uses **+**      | Formula uses **×**     |

#### 1. Classical Probability:

Classical probability is used when all possible outcomes are equally likely (same chance).

Classical probability म्हणजे सर्व outcomes equally likely (सारखी chance) असतात तेव्हा वापरली जाणारी probability.

Formula:

        P(A) = Favorable outcomes / Total outcomes
  ​
Where:

- P(A) = Probability of event A  
- Favorable outcomes = outcomes that satisfy the event  
- Total outcomes = all possible outcomes

##### - Example:

Coin Toss

Sample Space: 
 
S = {Head, Tail}  
Event A = Getting Head  
P(A) = 1 / 2

So, the probability of getting Head = 0.5 (50%).

#### 2. Empirical Probability (Experimental Probability)

Empirical probability is calculated using actual experiments or real data.  
Probability based on observed results from experiments.

Empirical probability म्हणजे actual experiment किंवा real data वरून calculate केलेली probability.

Formula:

	P(A) = Number of times event occurs / Total number of trials

##### - Example:

A coin is tossed 100 times.

Head appears 60 times.

P(Head)= 60 / 100	​

=0.6

So, the experimental probability of getting Head = 0.6.​


#### 3. Subjective Probability:

Subjective probability is based on personal judgement, belief, or experience rather than exact calculations.  
Probability based on human opinion or experience.

Subjective probability म्हणजे personal judgement, belief किंवा experience वर based probability.   
Exact calculation नसून व्यक्तीच्या experience किंवा opinion वर आधारित probability.

##### - Example:

Weather Forecast

A meteorologist says:

"Tomorrow there is a 70% chance of rain."

This prediction is based on past data, observation, and experience.

#### 4. Conditional Probability

Conditional Probability is the probability of an event A occurring given that another event B has already occurred.

In simple words:
It measures the probability of A happening when we already know B happened.

एक घटना झाल्यावर दुसऱ्या घटनेची probability

Formula: 

P(A∣B)= P(A∩B)P / (B)  

​
Where:

P(A|B) = Probability of A given B  
P(A ∩ B) = Probability that both A and B occur  
P(B) = Probability of event B

- ##### Example 1 :

Deck मध्ये 52 cards आहेत.   
Event A = card is King   
Event B = card is Face card Face cards = 12 (J, Q, K)  
Kings = 4

P(A∣B) = 4/12 ​= 1​/3

- ##### Example 2:

Class मध्ये students

Event A = pass math  
Event B = pass physics

👉 Physics pass झालेल्या students मधून math pass probability काढतो.


#### 5. Joint Probability

Joint Probability is the probability that two events occur together at the same time.  
It measures the probability of event A and event B happening together.

Formula :

P(A∩B)

Example

Two coins are tossed.

Sample Space:

S = {HH, HT, TH, TT}

Event A = First coin is Head  
Event B = Second coin is Head

Joint probability:

P(A∩B) = P(HH) = 1 / 4

So, the probability that both coins show Head = 1/4.

#### 6. Marginal Probability

Marginal Probability is the probability of a single event occurring without considering other events.  
It is the probability of one event independently.

Example:

Two coins are tossed.

Sample Space:

S = {HH, HT, TH, TT}

Event A = First coin is Head

Possible outcomes for A:

HH, HT

P(A)= 2/4 = 1/2

So, the probability that the first coin is Head = 1/2.

- ## Bayes Theorem:

-Bayes’ Theorem is a fundamental concept in probability theory that describes how to update the probability of an event when new information (evidence) becomes available.  
-It was introduced by Thomas Bayes.

#### Basic Idea

👉 आपल्याकडे आधी काही belief / probability असते (prior)  
👉 नंतर आपल्याला new evidence मिळते  
👉 त्या evidence च्या आधारे आपण probability update करतो (posterior)

#### Formula:

P(A∣B) = P(B∣A) × P(A) / P(B)

#### 1. Based on Conditional Probability

Bayes’ theorem is derived from conditional probability:

P(A∣B) = P(A∩B) / P(B)

P(B∣A) = P(A∩B) / P(A)

Combining both:

P(A∣B)= P(B∣A) × P(A) / P(B)​

#### 2. Important Components

Prior Probability (P(A))  
👉 A बद्दल आधीची belief  
(Marathi: आधीची माहिती)

Likelihood (P(B|A))  
👉 A true असेल तर B होण्याची probability

Evidence (P(B))  
👉 B होण्याची overall probability

Posterior Probability (P(A|B))  
👉 B झाल्यावर A true असण्याची updated probability

```python
# Import libraries
from sklearn.feature_extraction.text import CountVectorizer  # CountVectorizer -> converts text (words) into numbers
from sklearn.naive_bayes import MultinomialNB # MultinomialNB -> Naive Bayes model used for text classification

# Sample data
emails = [
    "free money now",
    "win cash prize",
    "meeting tomorrow",
    "project discussion"
]

labels = [1, 1, 0, 0]  # 1 = Spam, 0 = Not Spam

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test data
test = ["free prize"]
test_data = vectorizer.transform(test)

# Prediction
prediction = model.predict(test_data)

print("Prediction:", prediction)
```

#### 1. Gaussian Naive Bayes (Continuous Data)

👉 Use:

Numeric data (marks, height, etc.)

#### Explanation:

GaussianNB → continuous data handle करतो

Model learns:

Pass students marks distribution  
Fail students marks distribution  
Input 80 → decide pass or fail

```python
from sklearn.naive_bayes import GaussianNB
import numpy as np
import matplotlib.pyplot as plt

# Features (marks)
X = np.array([[85], [90], [78], [60], [65], [70]])

# Labels (1 = Pass, 0 = Fail)
y = np.array([1, 1, 1, 0, 0, 0])

# Model
model = GaussianNB()
model.fit(X, y)

# Prediction
prediction = model.predict([[80]])
print("Prediction for 80 marks:", prediction)

# ---------------- GRAPH ----------------

# Separate data for plotting
pass_marks = X[y == 1]
fail_marks = X[y == 0]

# Plot points
plt.scatter(pass_marks, [1]*len(pass_marks), label="Pass", marker='o')
plt.scatter(fail_marks, [0]*len(fail_marks), label="Fail", marker='x')

# Plot predicted point
plt.scatter(80, prediction, label="Prediction (80)", marker='*', s=200)

# Labels
plt.xlabel("Marks")
plt.ylabel("Class (0=Fail, 1=Pass)")
plt.title("Gaussian Naive Bayes Classification")

plt.legend()
plt.show()
```

#### 2. Multinomial Naive Bayes (Text Data)

👉 Use:

Text classification (most common)

#### Explanation:

Text → numbers (CountVectorizer)  
MultinomialNB → word frequency वापरतो  
“free prize” → spam probability जास्त

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import matplotlib.pyplot as plt
import numpy as np

emails = ["free money", "win prize", "meeting tomorrow", "project work"]
labels = [1, 1, 0, 0]  # 1 = Spam, 0 = Not Spam

# Convert text to numbers
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(emails)

# Train model
model = MultinomialNB()
model.fit(X, labels)

# Test data
test = vectorizer.transform(["prize tomorrow"])
prediction = model.predict(test)

print("Prediction:", prediction)

# ---------------- GRAPH ----------------

words = vectorizer.get_feature_names_out()
X_array = X.toarray()

spam_counts = X_array[np.array(labels) == 1].sum(axis=0)
ham_counts = X_array[np.array(labels) == 0].sum(axis=0)

x = np.arange(len(words))

plt.bar(x - 0.2, spam_counts, width=0.4, label="Spam")
plt.bar(x + 0.2, ham_counts, width=0.4, label="Not Spam")

plt.xticks(x, words, rotation=45)
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.title("Word Frequency in Spam vs Not Spam")

plt.legend()
plt.tight_layout()
plt.show()
```
