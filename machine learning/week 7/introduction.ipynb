#### What is Machine learning:

-Machine learning is a branch of Artificial Intelligence that focuses on developing models and algorithms that let computers learn from data without being explicitly programmed for every task.  

-In simple words, ML teaches systems to think and understand like humans by learning from the data.

-Machine Learning is a field of Artificial Intelligence that enables systems to learn from data, identify patterns, and make decisions with minimal human intervention.

-Example:
Machine Learning is used in email spam filtering to decide
👉 Spam email or Normal (Inbox) email

- Why used machine learning:

1️⃣ Too Much Data

-Today we have huge amounts of data  
-Humans cannot analyze it manually  
-Machine Learning can handle big data easily

📌 Example:  
Millions of YouTube videos → ML recommends videos you like

---------------------------------------------------------
2️⃣ No Fixed Rules Possible

-Some problems cannot be solved with fixed rules  
-Patterns keep changing

📌 Example:  
Spam emails change words every day → ML learns new spam patterns

---------------------------------------------------------
4️⃣ Saves Time and Effort

Manual work is slow and costly  
ML automates decision making

📌 Example:  
Face unlock is faster than typing a password

--------------------------------------------------------
5️⃣ Gives Personalized Results

Different users get different results  
ML understands user behavior

📌 Example:  
Netflix suggests movies based on your interest

- Difference between AI and ML:

| Point           | Artificial Intelligence (AI)          | Machine Learning (ML)           |
| --------------- | ------------------------------------- | ------------------------------- |
| Learning        | Works on predefined rules             | Learns from data                |
| Rules / Data    | Humans write rules                    | Machine creates rules from data |
| Adaptability    | Needs rule changes for new situations | Adapts automatically            |
| Improvement     | Performance stays the same            | Performance improves over time  |
| Personalization | Same behavior for all users           | Personalized behavior           |



#### Types of Machine Learning:

#### 1️⃣ Supervised Learning

-In Supervised Learning, the machine learns using labeled data  
(Labeled = input + correct output is already given)

-Example:

       Data              output
|----------------||------------------|
iq   |   cgpa    |    eligible or not|
-----|-----------|-------------------|
89   |   8.2     |    y              |
78   |   7.9     |    y              |
67   |   6.5     |    N              |



- Types of Supervised Learning:

Supervised learning algorithms are generally categorized into two main types: 

a.Classification   
b.Regression 


a.Classification:  
-where the goal is to predict discrete labels or categories

-Classification मध्ये output fixed category / label असतो.
म्हणजे answer “कसल्या प्रकारचा आहे?” हे सांगतो.

ex:  
pass/fail  
spam/not span  

Marks  |	Result
-------|----------------
85	|       Pass
30     |	Fail


Types of Classification:

A. Binary Classification → 2 classes

Ex:   
Yes / No  
Pass / Fail

Binary classification has only two possible classes.  
The machine chooses one out of two options.

-How it works:

       Training data contains two labels  
       The model learns patterns from the data  
       For new input, it predicts one of the two classes

B. Multi-class Classification → जास्त classes

Multi-class classification has more than two classes.  
The machine chooses one class from many options.

Ex:  
Grade A / B / C  
Cat / Dog / Bird

-How it works:

       Training data contains multiple labels  
       The model learns patterns for each class  
       For new input, it predicts the most suitable class




b. Regression: 
-where the aim is to predict continuous numerical values

-A continuous value is a number that can take any value within a range.

Examples:  
Marks: 72, 85, 91.5  
House price: ₹4,50,000  
Temperature: 28.3°C  
Salary: 35,000

-How it works:

1.Collect training data with input features and numerical output values.  
2️.Feed this data to the model for training.  
3️.The model learns the relationship between input and output.  
4️.A new input value is given to the trained model.  
5️.The model predicts a continuous numerical value as the output.

- Types of Regression:

1️⃣ Linear Regression

Relationship is a straight line  
Most commonly used regression method

2️⃣ Multiple Linear Regression

Uses more than one input variable  
Example: Experience + Education → Salary

3️⃣ Polynomial Regression

Relationship is curved  
Used when data is not linear

---------------------------------------------------------------------------------------------------------

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Input data
X = np.array([[1], [2], [3], [4], [5]])

# Output data
y = np.array([10, 20, 30, 40, 50])

# Model
model = LinearRegression()

# Training
model.fit(X, y)

# Prediction
pred = model.predict([[6]])

print("Prediction:", pred)
```

#### 2️⃣ Unsupervised Learning:

In unsupervised learning, the machine is given only input data.  
There is no teacher and no predefined answers.  
The model tries to find hidden patterns, structures, or relationships in the data by itself.  

-How it Works -

1️⃣ Collect unlabeled data (only inputs).  
2️⃣ Feed the data to the algorithm.  
3️⃣ The model analyzes similarities and differences.  
4️⃣ It groups or organizes data based on patterns.  
5️⃣ The discovered patterns are used for analysis or decision-making.

Example:

Image Grouping:-

Machine ला खूप images दिल्या:  
Cats  
Dogs  
Birds  
पण image ला नाव दिलेलं नाही

-Machine:  
Similar features असलेल्या images group करते

➡️ Same type images एकत्र येतात

-Key Points:  

No labeled data  
No predefined output  
Finds hidden patterns  
Used for exploration and discovery


- Common Techniques in Unsupervised Machine Learning:

1️. Clustering  
2️. Association  
3️. Dimensionality Reduction

1️. Clustering:

Clustering means grouping similar data points together.

Machine स्वतःच ठरवते  
“हे data points एकसारखे आहेत, म्हणून एकाच group मध्ये ठेवूया.”

-Common Algorithm:  
    K-means clustering

--------------------------------------------------------------------

2️. Association:

Association means finding relationships between items in large datasets.

Machine शोधते  
“कोणत्या गोष्टी एकत्र जास्त वेळा येतात?”

Example:

Supermarket data:  
Milk + Bread  
Diapers + Baby food

👉 Pattern:
“Milk घेणारे customers बहुतेक Bread पण घेतात”

-Common Algorithm:  
Apriori algorithm

-------------------------------------------------------------

3️. Dimensionality Reduction:

Dimensionality reduction means reducing the number of features while keeping important information.

Machine unnecessary details काढून टाकते  
आणि important data ठेवते.

Example:

Student data:  
Height, Weight, Age, Attendance, Marks  

👉 Machine important features ठेवते  
👉 Unimportant किंवा repeated features reduce करते

-Common Algorithm:  
    PCA (Principal Component Analysis)

```python
from sklearn.cluster import KMeans
import numpy as np

# Data
X = np.array([
    [1, 2],
    [1, 4],
    [1, 0],
    [10, 2],
    [10, 4],
    [10, 0]
])

# Model
model = KMeans(n_clusters=2) 

# Training
model.fit(X)

# Cluster labels
print(model.labels_)
```

#### 3️⃣ Semi-Supervised Learning:

Semi-supervised machine learning is a type of machine learning that uses both labeled and unlabeled data for training.

In semi-supervised learning:

Few data points are labeled  
Most of the data is unlabeled

The model first learns from the labeled data and then uses patterns in unlabeled data to improve learning

-In simple words:  
Machine ला थोडा teacher असतो, बाकी ती स्वतः शिकते.

-How It Works -

1️⃣ Collect a large dataset (mostly unlabeled).  
2️⃣ Label a small portion of the data.  
3️⃣ Train the model using labeled data.  
4️⃣ The model finds patterns in unlabeled data.  
5️⃣ Both are combined to make better predictions.

Example

👉 Image dataset:

1000 images  
Only 100 images are labeled (Cat / Dog)

🧠 Model:

Learns from 100 labeled images  
Uses similarity to classify remaining images

-Why Use Semi-Supervised Learning?


Labeling data is expensive and time-consuming  
Large unlabeled data is easily available  
Gives better accuracy than using small labeled data alone

#### 4️⃣ Reinforcement Learning:

Reinforcement Learning is a type of machine learning where a machine learns by trying actions and learning from feedback.

The machine is not told the correct answer.  
It is only told whether its action is good or bad.

Step-by-Step :

1️⃣ Machine काहीतरी action करते  
2️⃣ Environment response देते  
3️⃣ If action correct → reward  
4️⃣ If action wrong → penalty  
5️⃣ Machine लक्षात ठेवते आणि next time better action घेते

Example:

How learning happens:

1️⃣ Dog sits → gets a biscuit (reward)  
2️⃣ Dog stands → no biscuit (penalty)  
3️⃣ Dog tries again  
4️⃣ Over time, dog learns: “If I sit, I get a reward”  

👉 This is Reinforcement Learning.

```python
import random

actions = ["left", "right"]

for i in range(5):
    action = random.choice(actions)

    if action == "right":
        reward = 1
    else:
        reward = -1

    print("Action:", action, "Reward:", reward)
```
